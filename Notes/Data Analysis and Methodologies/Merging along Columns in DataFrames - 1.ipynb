```python
import pandas as pd
```

```python
excelfile = pd.ExcelFile("file.xlsx")
df_continent = excelfile.parse("Sheet1")
df_country = excelfile.parse("Sheet2")
```

```python
df_continent
```

```python
df_country
```

```python
#joining tables(inner,left,right,full)
pd.merge(df_continent, df_country, on='continentID', how='left')
```

```python
#merging on multiple columns
pd.merge(df_continent, df_country, on=['continentID','var'], how='left')
```

```python
#using suffixes
pd.merge(df_continent, df_country, on='continentID', how='left', suffixes=('_continent', '_country'))
```
