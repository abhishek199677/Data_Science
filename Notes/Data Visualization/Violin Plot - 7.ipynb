```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#violin plot - combination of box and kde plot
```

```python
ds = np.random.randn(80)
```

```python
sns.violinplot(ds)
plt.savefig("7.1.png")
```

```python
sns.violinplot(ds, bw=0.2)
plt.savefig("7.2.png")
```

```python
sns.violinplot(ds, inner='stick')
plt.savefig("7.3.png")
```
