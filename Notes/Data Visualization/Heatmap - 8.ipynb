```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("flights")
```

```python
df.head()
```

```python
df2 = df.pivot("year", "month", "passengers")
df2
```

```python
sns.heatmap(df2)
plt.savefig("8.1.png")

#divergent colormap - more frequency, light color
#convergent - vice versa
```

```python
sns.heatmap(df2, annot=True, fmt='d')
plt.savefig("8.2.png")
```

```python
#local centre - can be used to predict profit or loss
sns.heatmap(df2, center=df2.loc[1955,'January'])
plt.savefig("8.3.png")
```
