```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
mean = [1,0]
covariance = [[1,0], [0,100]]
```

```python
ds = np.random.multivariate_normal(mean, covariance, 500)
```

```python
dframe = pd.DataFrame(ds, columns=['col1','col2'])
```

```python
sns.kdeplot(dframe)
plt.savefig("kde2_image1.png")
```

```python
sns.kdeplot(dframe, shade=True)
plt.savefig("kde2_image2.png")
```

```python
sns.kdeplot(dframe, bw=1.1)
plt.savefig("kde2_image3.png")
```

```python
sns.kdeplot(dframe, bw='silverman')
plt.savefig("kde2_image4.png")
```

```python
sns.jointplot('col1', 'col2', dframe, kind='kde')
plt.savefig("kde2_image5.png")
```
