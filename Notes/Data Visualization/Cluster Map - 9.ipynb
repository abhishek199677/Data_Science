```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("flights")
```

```python
df.head()
```

```python
df2 = df.pivot('year', 'month', 'passengers')
df2
```

```python
sns.clustermap(df2)
plt.savefig("9.1.png")
```

```python
sns.clustermap(df2, col_cluster=False)
plt.savefig("9.2.png")
```

```python
sns.clustermap(df2, standard_scale=0)   #standardize by row
plt.savefig('9.3.png')
```

```python
sns.clustermap(df2, standard_scale=1)   #standardize by column
plt.savefig('9.4.png')
```
