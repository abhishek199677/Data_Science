```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("diamonds").sample(frac=1).head(100)
```

```python
# df.head()
```

```python
df.shape
```

```python
sns.lmplot("price", "carat", df)
plt.savefig("6.1.png")
```

```python
sns.lmplot("price", "carat", df,
           scatter_kws={'marker':'x', 'color':'green'},
           line_kws={'linewidth':1, 'color':'red'})
plt.savefig("6.2.png")
```

```python
#higher order trendlines
sns.lmplot("price", "carat", df, order=4,
           scatter_kws={'marker':'x', 'color':'green'},
           line_kws={'linewidth':1, 'color':'red'})
plt.savefig("6.3.png")
```

```python
#scatter plot without trendline
sns.lmplot("price", "carat", df, fit_reg=False)
plt.savefig("6.4.png")
```

```python
#hue parameter
sns.lmplot("price", "carat", df, hue="cut", markers=['^', 'v', '*', '.', 's'])
plt.savefig("6.5.png")
```

```python
sns.lmplot("price", "carat", df, hue="cut")
plt.savefig("6.6.png")
```

```python
#local regression
sns.lmplot("price", "carat", df, lowess=True)
plt.savefig("6.7.png")
```

```python
#regplot
sns.regplot("price", "carat", df)
plt.savefig("6.8.png")
```

```python
#subplots
image, (plt1,plt2) = plt.subplot(1,2,sharey=True)
sns.regplot("price", "carat", df, ax=plt1)
sns.boxplot("carat", "depth", df, color='green', ax=plt2)
plt.savefig("6.9.png")
```
