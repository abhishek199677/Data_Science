```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds1 = np.random.randn(1000)
ds2 = np.random.randn(100)
```

```python
plt.hist(ds1)              #by default histogram has 10 bins
plt.savefig("image1.png")
plt.show()
```

```python
plt.hist(ds2)
plt.savefig("image2.png")
plt.show()
```

```python
#combining the ds1 and ds2 in same histogram
plt.hist(ds1, density=True, color='green', bins=15, alpha=0.5)
plt.hist(ds2, density=True, bins=15, alpha=0.5)
plt.savefig("image3.png")
plt.show()
```

```python
#using seaborn
ds3 = np.random.randn(1000)
ds4 = np.random.randn(1000)
```

```python
sns_plot = sns.jointplot(ds3, ds4)
sns_plot.savefig("image4.png")
```

```python
sns_plot2 = sns.jointplot(ds3, ds4, kind='hex')
sns_plot2.savefig("image5.png")
```
