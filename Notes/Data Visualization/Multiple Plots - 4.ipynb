```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds = np.random.randn(100)
```

```python
#kde and hist
sns.distplot(ds, bins=15)
plt.savefig("4.1.png")
```

```python
#kde and rug
sns.distplot(ds, bins=10, hist=False, rug=True)
plt.savefig("4.2.png")
```

```python
sns.distplot(ds, bins=10,
             kde_kws={'color':'red', 'label':'KDE Graph'},
             hist_kws={'color':'green', 'label':'Histogram', 'alpha':0.5})
plt.savefig("4.3.png")
```

```python
s1 = pd.Series(ds, name='s1')
```

```python
sns.distplot(s1, bins=15)
plt.savefig("4.4.png")
```
