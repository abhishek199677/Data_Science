```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy 
```

```python
ds = np.random.randn(100)
```

```python
sns.rugplot(ds)
plt.hist(ds, alpha=0.5)
plt.savefig("kde_image1.png")
```

```python
x_axes = np.linspace(ds.min()-1, ds.max()+1, 50)
```

```python
bandwidth = ((4*ds.std()**5)/(3*len(ds))) ** 0.2
```

```python
kernels=[]

for point in ds:
    kernel = scipy.stats.norm(point, bandwidth).pdf(x_axes)
    kernels.append(kernel)
    
    kernel = kernel/kernel.max()
    kernel = kernel*0.6
    
    plt.plot(x_axes, kernel, alpha=0.5, color="red")
    

plt.savefig("kde_image2.png")
#will give guassian curve
```

```python
kde = np.sum(kernels, axis=0)
kde_fig = plt.plot(x_axes, kde, color="green")
sns.rugplot(ds)
plt.suptitle("KDE Plot")
plt.savefig("kde_image3.png")
#will give KDE Plot
```
